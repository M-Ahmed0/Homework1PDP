```python
import re
```

```python
def clean_comic_data(filename):
    cleaned_data = []
    
    with open(filename, 'r', encoding='utf-8', errors='replace') as file:
        for line in file:
            line = line.strip() # Remove leading/trailing whitespace from the line
            if line:
                try:
                        # Regular expression match to extract information from the line
                    match = re.match(r"([A-Z]) - ([A-Za-z0-9\s]+) - ([A-Z]?\d+) - ([A-Z][\w\s\d()Α-Ωα-ω]*)\.cbr", line)
                    
                    # Extract the column names
                    if match:
                        field_name = match.group(1)
                        series_name = match.group(2)
                        book_number = match.group(3)
                        book_title = match.group(4)
                        cleaned_data.append([field_name, series_name, book_title, book_number])
                    else:
                        print("Pattern matching failed for line:", line)
                except Exception as e:
                    print("Error processing line:", line)
                    print("Error message:", str(e))
    
    return cleaned_data # Return the cleaned data as a list

# Specify the path to your comics.txt file
comics_file = "comics.txt"

# Call the clean_comic_data function
cleaned_comics = clean_comic_data(comics_file)

# Specify the path for the output text file
output_file = "cleaned_comics.txt"

# Write the cleaned data to the text file with comma separation
with open(output_file, 'w', encoding='utf-8') as file:
    file.write("FieldName,Series Name,Book Title,Book Number\n")  # Write header
    for comic in cleaned_comics:
        line = ','.join(comic) + '\n'
        file.write(line)

print("Data saved to", output_file)
```
